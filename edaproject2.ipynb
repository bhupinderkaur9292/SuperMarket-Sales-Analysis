```python
nan_df=df[df.isna().any(axis=1)]
nan_df.head()
```

```python
df.dropna(how='all',inplace=True)
df.isna().sum()
```

```python
df.duplicated().value_counts()
```

```python
dfa=df.drop_duplicates(subset=None, keep='first', inplace=False)
dfa
```

```python
dfa.columns=['Order_ID', 'Product', 'Quantity_Ordered', 'Price_Each', 'Order_Date', 'Purchase_Date']
dfa.head()
```

```python
dfa['Date']=dfa['Order_Date'].str[0:8]
dfa
```

```python
dfa['Months']=dfa['Order_Date'].str[0:2]
dfa
```

```python
dfa[['House_No.', 'City', 'Pin_Code']] = dfa.Purchase_Date.str.split(",", expand=True)
dfa
```

```python
dfa.Months.value_counts()
```

```python
dfa=dfa[dfa['Months'].str[:2] != 'Or']
dfa.head()
```

```python
dfa.shape
```

```python
dfa['Months']=dfa['Months'].astype('int64')
dfa['Order_Date']=pd.to_datetime(dfa['Order_Date'])
dfa['Date']=pd.to_datetime(dfa['Date'])
dfa.info()
```

```python
dfa.Order_ID.value_counts()
```

```python
dfa.Quantity_Ordered.value_counts()
```

```python
dfa['Quantity_Ordered']=dfa['Quantity_Ordered'].astype('int32')
dfa['Price_Each']=dfa['Price_Each'].astype('Float64')
dfa.info()
```

```python
dfa['Sales']=dfa['Quantity_Ordered']*dfa['Price_Each']
dfa
```

```python
Monthly_sales=dfa.groupby('Months').Sales.sum()
Monthly_sales
```

```python
City_sales=dfa.City.value_counts()
City_sales
```

```python
plt.figure(figsize=(16,6))
sns.countplot(x=dfa['City'])
plt.xlabel("City Name")
plt.ylabel("Number of sells")
plt.title("City vs sales")
```

```python
dfa['Hour']=dfa.Order_Date.dt.hour
```

```python
dfa.groupby('Hour').count()
```

```python
hours=[hour for hour, dfa in dfa.groupby('Hour')]
plt.plot(hours, dfa.groupby('Hour').count())
plt.grid(True)
plt.xticks(hours)
plt.xlabel('Hours in 24 Hr Format')
plt.ylabel('Orders')
plt.show()
```

```python
Highest_Orders=dfa.Product.value_counts()
Highest_Sales=dfa.groupby('Product').Sales.sum().sort_values(ascending=False)
fig, axes=plt.subplots(1,2, figsize=(20,7))
sns.barplot(ax=axes[0], x=Highest_Orders.index, y=Highest_Orders.values)
axes[0].set_xlabel('Product')
axes[0].set_ylabel('No. of Orders')
axes[0].tick_params(axis='x', rotation=90)
sns.barplot(ax=axes[1],x=Highest_Sales.index, y=Highest_Sales.values)
axes[1].set_xlabel('Product')
axes[1].set_ylabel('Sales Amount')
axes[1].tick_params(axis='x', rotation=90)
plt.show()
```
